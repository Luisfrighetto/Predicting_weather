### Treino do modelo

Modelo de regressão de crista (Ridge regression) que penaliza os coeficientes para levar em conta a multicolinearidade

```python
from sklearn.linear_model import Ridge
```

```python
#Verificação de correlação para multicolinearidade
df.corr()
```

```python
## Ridge ajuda com a multicolinearidade

rr = Ridge(alpha=.1) #parametro que controla quanto os coef sao reduzidoss
```

```python
preditores = df.columns[~df.columns.isin(["target","name","station"])]
```

```python
preditores
```

Temos uma serie de dados temporal, o normal ao tentar estimar um erro de um modelo de aprendizado de maquina é usar VALIDAÇÃO CRUZADA. Porem quando temos uma serie temporal nao podemos usar dados do futuro para prever o passado. Usamos entao o recurso de BACKTESTING. Ou validacao cruzada de dados de serie temporal para levar em conta a natureza da serie temporal dos dados.

```python
def backtest(df, model, preditores, start=3650,step=90): #start representa a quantidade de dados que queremos analizar 
    all_predictions = []                                #step representa o intervalo de dados de 90 dias      
    
    for i in range(start, df.shape[0], step):
        train = df.iloc[:i,:]
        test = df.iloc[i:(i+1),:]
        
        model.fit(train[preditores], train["target"])
        
        preds = model.predict(test[preditores]) #gerar em numpy  array
        
        preds = pd.Series(preds, index=test.index) # trannsformar em series panda 
        
        combine = pd.concat([test["target"], preds], axis=1)
        
        combine.columns = ["atual","predicao"]
        
        combine["difd"] = (combine["predicao"] - combine["atual"]).abs() #abs é valores absolutos
        
        all_predictions.append(combine)
    return pd.concat(all_predictions, axis=0)   
    
    
    
    
```

```python
predictions = backtest(df,rr, preditores)
```

```python
predictions #note que pulamos os 10 primeiros anos 
```

```python
# metrica de acuracia: erro absoluto medio

from sklearn.metrics import mean_absolute_error

mean_absolute_error(predictions["atual"],predictions["predicao"])
```

Em media estamos 5 farenhight distantes do verdadeiro valor em temperatura
Vamos melhorar a acuracia criando novos preditores. medias de dias anteriores

### Adicionando preditores ao modelo

```python
def pct_diff(old,new):
    return (new - old) /old

def compute_rolling(df, horizonte, col):
    label = f"rolling_{horizonte}_{col}"
    
    df[label] = df[col].rolling(horizonte).mean()
    df[f"{label}_pct"] = pct_diff(df[label], df[col])
    return df

rolling_horizonte = [3,14]

for horizonte in rolling_horizonte:
    for col in["tmax","tmin","prcp"]:
        df = compute_rolling(df, horizonte, col)
```

```python
df
```

```python
df = df.iloc[14:,:]
```

```python
df=df.fillna(0)
```

```python
def expend_mean(df):
    return df.expanding(1).mean() #olha cada linha e pega a media de todas as linhas anteriores

#encontrando a temperatura media em um determinado mes, mas apenas em dias para os quais ja tinhamos informações 
for col in ["tmax","tmin","prcp"]:
    df[f"month_avg_{col}"] = df[col].groupby(df.index.month, group_keys=False).apply(expend_mean)
    df[f"day_avg_{col}"]  = df[col].groupby(df.index.day_of_year,  group_keys=False).apply(expend_mean)
```

```python
df
```

```python
preditores = df.columns[~df.columns.isin(["target","name","station"])]
```

```python
preditores
```

### Recriando o modelo com as novas variaveis

```python
predictions = backtest(df,rr, preditores)
```

```python
mean_absolute_error(predictions["atual"],predictions["predicao"])
```
